# Model 1

## Features

```python
coords = coords.drop(['geometry'], axis=1)
```

```python
df_time = pd.DataFrame()
df_time["month"] = rain.index.month
df_time["day"] = rain.index.day
df_time["hour"] = rain.index.hour
#df_time["year"] = rain.index.year
```

```python
X = coords.merge(df_time, how="cross")
y = rain.stack().reset_index()[0]
X
```

## Normalize data

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.MinMaxScaler()

X_norm = pd.DataFrame()
X_norm[X.columns] = scaler.fit_transform(X[X.columns])
X_norm.head()
```

## Regression algo

```python
m_samples, n_features = X.shape
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math
from torch import optim
```

#### Train and Test data

```python
#train data
X_train = X_norm.iloc[ : int(m_samples * 0.8)]
y_train = y[ : int(m_samples * 0.8)]

#test data
X_test = X_norm.iloc[int(m_samples * 0.8) : ]
y_test = y[int(m_samples * 0.8) : ]

X_train = torch.tensor(X_train.values, dtype=torch.float)
y_train = torch.tensor(y_train.values, dtype=torch.float)
X_test = torch.tensor(X_test.values, dtype=torch.float)
y_test = torch.tensor(y_test.values, dtype=torch.float)
```

#### Modele

```python
torch.manual_seed(0)

class rain_module(nn.Module):
    def __init__(self):
        super().__init__()
        #self.weights = nn.Parameter(torch.randn(n_features, dtype=torch.float, requires_grad=True) / math.sqrt(n_features))
        #self.bias = nn.Parameter(torch.randn(1, dtype=torch.float, requires_grad=True))
        self.linear_layer = nn.Linear(in_features=n_features, out_features=1)
        
    def forward(self, xb: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(xb) #xb @ self.weights + self.bias 
```

```python
model = rain_module()

loss_func = nn.MSELoss()

optimizer = optim.SGD(model.parameters(), lr=0.000001)
```

```python

# Setup device agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

model.to(device)
next(model.parameters()).device
```

```python
epochs = 2

bs = 64 #Pourquoi on train le model avec plein de petit batch size au lieu d un gros batch

def fit():
    
    train_loss_evo = []
    test_loss_evo = []
    for epoch in range(epochs):
        for i in range((m_samples - 1) // bs + 1):
            start_i = i * bs
            end_i = start_i + bs
            xb = X_train[start_i:end_i]
            yb = y_train[start_i:end_i]
            pred = model(xb)
            loss = loss_func(pred, yb)
            
            optimizer.zero_grad()

            loss.backward() #Comment loss.backward modifie les parametres?
            
            optimizer.step()

            if i % bs ==0:
                test_loss_evo.append(loss.detach().numpy())
        
        ##TESTING
        #model.eval()
        #with torch.inference_mode(): 
        #    test_pred = model_0(X_test)
        #    test_loss = loss_func(test_pred, y_test)
    return test_loss_evo
loss_evo = fit()
```

```python
#Evolution de la fonction cout
plt.plot(np.arange(len(loss_evo)), loss_evo)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(np.arange(len(y_train)), y_train.detach().numpy(), c="b", label="Correct values")
plt.scatter(np.arange(len(y_train)), model(X_train).detach().numpy(), c="r", label="Correct values")
plt.show()
```

# Model 2 : Regression lineaire

#### Prediction of next hour on jauge number 7

```python
plt.figure(figsize=(15,8))
plt.plot(rain.index, rain[7])
plt.show()
```

```python
r_7 = pd.DataFrame(rain[7])
r_7["h_1"] = r_7[7].shift(+1)
r_7["h_2"] = r_7[7].shift(+2)
r_7["h_3"] = r_7[7].shift(+3)
r_7["h_4"] = r_7[7].shift(+4)
r_7["h_5"] = r_7[7].shift(+5)
r_7["h_6"] = r_7[7].shift(+6)
r_7
```

```python
y7 = np.array(r_7[7])
x7 = r_7.drop([7], axis=1).fillna(0).values #Do i have to fill na values
                                            #Sdhould i remove the 6 first row instead id fillna

x7_train = x7[:int(len(x7)*0.8)]
x7_test = x7[int(len(x7)*0.8):]
y7_train = y7[:int(len(y7)*0.8)]
y7_test = y7[int(len(y7)*0.8):]
```

## 1. Sklearn

```python
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
```

```python
lin_model.fit(x7_train, y7_train)
print("R carre =", lin_model.score(x7_train, y7_train))
pred = lin_model.predict(x7_test)

plt.figure(figsize=(10,5))
plt.title("Sklearn linear regression on model 7")
plt.plot(pred, label="pred")
plt.plot(y7_test, c="r", label="target")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(pred, y7_test)
```

## 2. Pytorch

```python
X7_train = torch.tensor(x7_train, dtype=torch.float)
Y7_train = torch.tensor(y7_train.reshape(-1,1), dtype=torch.float)
X7_test = torch.tensor(x7_test, dtype=torch.float)
Y7_test = torch.tensor(y7_test.reshape(-1,1), dtype=torch.float)
```

```python
torch.manual_seed(0)

class r7_module(nn.Module):
    def __init__(self):
        super().__init__()
        #self.weights = nn.Parameter(torch.randn(n_features, dtype=torch.float, requires_grad=True) / math.sqrt(n_features))
        #self.bias = nn.Parameter(torch.randn(1, dtype=torch.float, requires_grad=True))
        self.linear_layer = nn.Linear(in_features=X7_train.shape[1], out_features=1)
        
    def forward(self, xb: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(xb) #xb @ self.weights + self.bias 
```

```python
model7 = r7_module()

loss_func7 = nn.MSELoss()
optimizer7 = optim.SGD(model7.parameters(), lr=0.01)
```

```python
train_loss_evo = []
test_loss_evo = []
for epoch in range(100):
    pred = model7(X7_train)
    loss = loss_func7(pred, Y7_train)
    
    optimizer7.zero_grad()

    loss.backward() #Comment loss.backward modifie les parametres?
    
    optimizer7.step()

    train_loss_evo.append(loss.detach().numpy())
    
    #TESTING
    model7.eval()
    with torch.inference_mode(): 
        test_pred = model7(X7_test)
        test_loss = loss_func7(test_pred, Y7_test)
        
        test_loss_evo.append(test_loss.detach().numpy())
```

```python
plt.plot(test_loss_evo, c="r")
plt.plot(train_loss_evo)
```

```python
pred = model7(X7_test)
print(mean_squared_error(pred.detach().numpy(), Y7_test.detach().numpy()))

plt.figure(figsize=(10,5))
plt.title("Tensor linear regression on model 7")
plt.plot(pred.detach().numpy(), label="pred")
plt.plot(Y7_test.detach().numpy(), c="r", label="target")
plt.legend()
plt.show()
```

```python
scaler = preprocessing.MinMaxScaler()
rain_norm = pd.DataFrame(scaler.fit_transform(rain), columns=rain.columns)
```

```python

rain_shift = []
for i in rain_norm.columns:
    r_i = pd.DataFrame({"target":rain[i].values})
    
    r_i["h_1_j"+str(i)] = rain_norm[i].shift(+1)
    r_i["h_2_j"+str(i)] = rain_norm[i].shift(+2)
    r_i["h_3_j"+str(i)] = rain_norm[i].shift(+3)
    r_i["h_4_j"+str(i)] = rain_norm[i].shift(+4)
    r_i["h_5_j"+str(i)] = rain_norm[i].shift(+5)
    r_i["h_6_j"+str(i)] = rain_norm[i].shift(+6)
    
    #Drop na rows
    r_i.dropna(inplace=True)
    
    rain_shift.append(r_i)
```

```python
m_samples = rain_shift[0].shape[0]

X=[np.array([]) for _ in range(m_samples)]
y=[np.array([]) for _ in range(m_samples)]
for df in rain_shift:    
    for i in range(m_samples):
        X[i] = np.append(X[i], df.drop(["target"], axis=1).iloc[i])
        y[i] = np.append(y[i], df["target"].iloc[i])
```

```python
for sample in X:
    sample.reshape(-1,1)
for sample in y:
    sample.reshape(-1,1)
```

```python
X_train = X[:int(len(X)*0.8)]
X_test = X[int(len(X)*0.8):]
y_train = y[:int(len(y)*0.8)]
y_test = y[int(len(y)*0.8):]
```

## 1. Sklearn

```python
lin_model=LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
print("R carre =", lin_model.score(X_train, y_train))
pred = lin_model.predict(X_test)
```
